```python
from ckiptagger import data_utils, construct_dictionary, WS, POS, NER
path = './data'
ws = WS(path)
```

```python
import pandas as pd
```

```python
df = pd.read_excel('./TEST/debug_data/心理學實驗(一) .xlsx')
```

```python
column_idx = 43
```

```python
df.insert(loc=column_idx+1,column='wseg',value="")
```

```python
df.columns[40:50]
```

```python
tmp_txts = []
tmp_rows = []
for row in range(0, df.shape[0]):
    text = df.iloc[row,column_idx]
    if type(text)!=str:
        continue
    text = text.replace('\r\n', '')
    text = text.replace('\n', '')
    tmp_txts.append(text)
    tmp_rows.append(row)
seg_txts = ws(tmp_txts)
print('finish seg: ',len(seg_txts))
for row, text in zip(tmp_rows, seg_txts):
    text = ' '.join(text)
    df.ins[row,'wseg']=text 
```
